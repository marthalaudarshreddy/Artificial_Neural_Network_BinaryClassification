```python
import tensorflow as ts
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('Churn_Modelling.csv')
data
```

```python
inde=data.iloc[:,3:-1]#because the customer id is not a dependent feature
depe=data.iloc[:,-1]
inde
```

```python
#convert words into values
geo=pd.get_dummies(data['Geography'],dtype=int,drop_first=True)
gend=pd.get_dummies(data['Gender'],dtype=int,drop_first=True)
```

```python
inde=inde.drop(['Geography','Gender'],axis=1)
inde
```

```python

inde=pd.concat([inde,geo,gend],axis=1)
inde
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inde,depe,test_size=0.33, random_state=0)
x_test
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x_train=scale.fit_transform(x_train)
x_test=scale.transform(x_test)
```

```python
# creating ANN
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,ReLU
from tensorflow.keras.layers import Dropout
```

```python
model=Sequential()

model.add(Dense(units=11,activation='relu'))
model.add(Dense(units=10,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(units=7,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
early_sto=ts.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0.0001,
    patience=1,
    verbose=0,
    mode='auto',
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0
)
```

```python
model_total=model.fit(x_train,y_train,batch_size=20, epochs=100,callbacks=[early_sto])
```

```python
model_total.history.keys()
```

```python
new_predict=model.predict(x_test)
new_predict=(new_predict>=0.5)
new_predict = new_predict.astype(int)
new_predict 
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
matr=confusion_matrix(y_test,new_predict)
matr
```

```python
accura=accuracy_score(new_predict,y_test)
accura
```
